```python
import tensorflow
```

```python
from keras.utils import img_to_array,array_to_img,load_img
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
from keras.models import Sequential
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train=ImageDataGenerator(rescale=1./255)
test=ImageDataGenerator(rescale=1./255)
```

```python
train_data=train.flow_from_directory(r"C:\Users\USER\Desktop\MLProject\Training_dataset",target_size=(100,100),class_mode="binary")
test_data=test.flow_from_directory(r"C:\Users\USER\Desktop\MLProject\Testing_dataset",target_size=(100,100),class_mode="binary")
```

```python
classifier=Sequential()
classifier.add(Conv2D(32,3,3,input_shape=(100,100,3),activation="relu"))
classifier.add(MaxPooling2D(pool_size=(2,2)))
classifier.add(Flatten())
classifier.add(Dense(128,activation="relu"))
classifier.add(Dense(1,activation="sigmoid"))

classifier.compile(optimizer='adam',
              loss='BinaryCrossentropy',
              metrics=['accuracy'])

classifier.fit(train_data,epochs=50,validation_data=test_data)
```

```python
def predict(path,classifier):
    predict=load_img(path,target_size=(100,100))
    predict_modified=img_to_array(predict)
    predict_modified=predict_modified/255
    predict_modified=np.expand_dims(predict_modified,axis=0)
    final=classifier.predict(predict_modified)
    print(final)
```

```python
path=r"C:\Users\USER\Desktop\Deep_Learning_Codebasics\Prediction Images\puppy-1903313_1280.jpg"
predict(path,classifier)
```

```python
path=r"C:\Users\USER\Desktop\Deep_Learning_Codebasics\Prediction Images\cat-6664360_1280.jpg"
predict(path,classifier)
```
